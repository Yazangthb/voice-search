```python
import numpy as np

def cosine_similarity(vec1, vec2):
  dot_product = np.dot(vec1, vec2)
  norm1 = np.linalg.norm(vec1)
  norm2 = np.linalg.norm(vec2)
  return dot_product / (norm1 * norm2)
```

```python
import IPython
import Levenshtein

def calculate_levenshtein_distances(list1, list2):
    if len(list1) != len(list2):
        raise ValueError("Both lists must have the same length.")

    distances = []

    for text1, text2 in zip(list1, list2):
        distance = Levenshtein.distance(text1, text2)
        distances.append(distance)

    return distances
```

```python
SPEECH_FILE = '/content/cv-valid-train/cv-valid-train/sample-000005.mp3'
IPython.display.Audio(SPEECH_FILE)
```

```python
class GreedyCTCDecoder(torch.nn.Module):
    def __init__(self, labels, blank=0):
        super().__init__()
        self.labels = labels
        self.blank = blank

    def forward(self, emission: torch.Tensor) -> str:
        """Given a sequence emission over labels, get the best path string
        Args:
          emission (Tensor): Logit tensors. Shape `[num_seq, num_label]`.

        Returns:
          str: The resulting transcript
        """
        indices = torch.argmax(emission, dim=-1)  # [num_seq,]
        indices = torch.unique_consecutive(indices, dim=-1)
        indices = [i for i in indices if i != self.blank]
        return "".join([self.labels[i] for i in indices])
```

```python
import torchaudio

torch.random.manual_seed(0)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

bundle = torchaudio.pipelines.WAV2VEC2_ASR_BASE_960H

model = bundle.get_model().to(device)

def preprocess_transcript(transcript):
    return transcript.lower().replace('|', ' ').strip()

def predict_transcript(audio_file):
  waveform, sample_rate = torchaudio.load(audio_file)
  waveform = waveform.to(device)

  if sample_rate != bundle.sample_rate:
      waveform = torchaudio.functional.resample(waveform, sample_rate, bundle.sample_rate)

  with torch.inference_mode():
      emission, _ = model(waveform)

  decoder = GreedyCTCDecoder(labels=bundle.get_labels())
  transcript = decoder(emission[0])
  return transcript
```

```python
transcript = predict_transcript(SPEECH_FILE)
print(transcript)
preprocess_transcript(transcript)
```

```python
def calculate_distances(df):
    distances = []
    predicted_texts = []

    for index, row in df.iterrows():
        actual_text = row['text']
        predicted_text = predict_transcript('/content/cv-valid-train/'+ row['filename'])
        preprocessed_text = preprocess_transcript(predicted_text)
        distance = Levenshtein.distance(actual_text, preprocessed_text)
        distances.append(distance)
        predicted_texts.append(preprocessed_text)

    return predicted_texts, distances

df['predicted_text'], df['levenshtein_distance'] = calculate_distances(df)
```

```python
def normalize_distance(row):
    actual_length = len(row['text'])
    predicted_length = len(row['predicted_text'])
    max_length = max(actual_length, predicted_length)
    if max_length == 0:
        return 0
    return row['levenshtein_distance'] / max_length

df['normalized_levenshtein_distance'] = df.apply(normalize_distance, axis=1)

overall_normalized_distance = df['normalized_levenshtein_distance'].mean()

print(f"Overall Normalized Levenshtein Distance: {overall_normalized_distance}")
```
