```python
from googletrans import Translator
import re

def detect_russian_sentences(text):
    # Regular expression to match Russian sentences (based on Cyrillic characters)
    russian_pattern = re.compile(r'[а-яА-ЯёЁ]+[\.,!?\s]*')
    
    # Find all matches
    russian_sentences = russian_pattern.findall(text)
    
    return russian_sentences

def translate_russian_sentences(text):
    translator = Translator()
    
    # Detect and translate Russian sentences
    russian_sentences = detect_russian_sentences(text)
    translated_sentences = []
    
    for sentence in russian_sentences:
        translation = translator.translate(sentence, src='ru', dest='en')
        translated_sentences.append(translation.text)
    
    return translated_sentences

def translate_document(filepath):
    with open(filepath, 'r', encoding='utf-8') as file:
        document_text = file.read()
    
    translated_sentences = translate_russian_sentences(document_text)
    
    # Replace original Russian sentences with translated ones
    for original, translated in zip(detect_russian_sentences(document_text), translated_sentences):
        document_text = document_text.replace(original, translated, 1)
    
    return document_text

# Example usage:
if __name__ == "__main__":
    translated_document = translate_document('email_8.txt')
    print(translated_document)
```
